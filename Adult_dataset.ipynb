```python
import pandas as pd
```

```python
data = pd.read_csv("adult.csv")
```

```python
data
```

```python
dataset = data.iloc[:,[0, 1, 3, 4, 6, 10, 11, 12, 14]]
```

```python
dataset
```

```python
for i in dataset.columns:
    if dataset[i].dtype=='object':
        dataset[i] = dataset[i].str.strip()
```

```python
dataset['salary'] = dataset['salary'].replace('>50K',1)
```

```python
l=[]
for i in dataset.columns:
    for j in range(dataset.shape[0]):
        if dataset[i][j]=='?':
            l.append(j)
dataset.drop(index=l,inplace=True)
```

```python
dataset.drop(['capital-loss', "capital-gain"], axis=1, inplace=True)
```

```python
dataset['workclass'].value_counts()
```

```python
dataset
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

for i in dataset:
    dataset[i] = label_encoder.fit_transform(dataset[i])
```

```python
dataset
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:,-1].values
```

```python
x.shape, y.shape
```

```python
y = y.reshape(y.shape[0], 1)
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=7,test_size=0.2)
```

```python
x_train, y_train
```

```python
models = {}

from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

from sklearn.svm import SVC
models['RBF SVC'] = SVC(kernel='rbf')
models['Sigmoid SVC'] = SVC(kernel='sigmoid')
models['Polynomial SVC'] = SVC(kernel='poly')

from sklearn.tree import DecisionTreeClassifier
models['Decision Tree'] = DecisionTreeClassifier()

from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier(n_estimators = 100)
```

```python
from sklearn.metrics import accuracy_score

accuracy = {}
model_score = {}

for key in models.keys():
    # Fit the classifier
    models[key].fit(x_train, y_train)
    # Make predictions
    predictions = models[key].predict(x_test)
    # model score
    model_score[key] = models[key].score(x_train, y_train)
    # Calculate metrics
    accuracy[key] = accuracy_score(predictions, y_test)
```

```python
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Model Score'])
df_model['Accuracy'] = accuracy.values()
df_model['Model Score'] = model_score.values()
df_model
```
